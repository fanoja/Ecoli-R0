# Tutorial: Create a conda environment from file


Step-by-step:
- Type: module load anaconda
- Type: conda env create --name ecoli-elfi --file environment.yml
- If ipykernel is not installed: conda install -c anaconda ipykernel
- To make the environment available as a Jupyter kernel: python -m ipykernel install --user --name=ecoli-elfi

Run the cells below to execute these steps.


Further reading:
- https://scicomp.aalto.fi/triton/apps/jupyter/
- https://elfi.readthedocs.io/en/latest/installation.html#

If the installation of environment.yml fails due to `pyhthon-graphviz`, remove it from the environment.yml file and install manually. So remove this under `pip`:
    `- python-graphviz==0.20.1`
   - To install manually: `pip install graphviz`
    

If the installation fails due to **legacy-install-failure**  during installation of elfi:
- This seems to be because `gcc` is missing
- simply `conda deactivate` and then `module load gcc`
- And after that, install the conda environment from `environment.yml` as detailed above.
